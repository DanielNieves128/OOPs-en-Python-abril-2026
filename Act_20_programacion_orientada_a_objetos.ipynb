lenguaje Python, crear la clase ¨animal¨ con los atributos "nombre", "raza" crear el constructor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definición de la clase
class Animal:
    # El constructor __init__ inicializa los atributos del objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función para que el animal ladre
    def ladrar(self):
        print(f"{self.nombre} está ladrando: ¡Guau guau!")

    # Función para que el animal corra
    def correr(self):
        print(f"{self.nombre}, el {self.raza}, está corriendo velozmente. 🐶")

    def morder(self):
      print(f"{self.nombre} esta mordiendo a la vecina 🤦🏽‍♂️ ")
# --- Creación de los objetos (instancias) ---

# Objeto 1: Pitbull
pitbull = Animal("Rocky", "Pitbull")
# creamos el objetivo perro

# Objeto 2: Chihuahua
chihuahua = Animal("Chispa", "Chihuahua")
# creamos el objetivo perro callejero con datos
perro_callejero = Animal("Bobby", "mestizo")
# --- Uso de las funciones ---

print("--- Interacción con el Pitbull ---")
pitbull.ladrar()
pitbull.correr()

print("\n--- Interacción con el Chihuahua ---")
chihuahua.ladrar()
chihuahua.correr()

print("\n--- Interacción con el Perro Callejero ---")
perro_callejero.ladrar()
perro_callejero.correr()
perro_callejero.morder()

print("Programa creado por Daniel Nieves NL=41")
```
